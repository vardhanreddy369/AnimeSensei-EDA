```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

anime_data = pd.read_csv('anime.csv')

print(anime_data.head())

print(anime_data.describe())


print(anime_data.isnull().sum())


sns.histplot(anime_data['rating'], bins=20, kde=True)
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Anime Ratings')
plt.show()

sns.countplot(x='type', data=anime_data)
plt.xlabel('Anime Type')
plt.ylabel('Count')
plt.title('Distribution of Anime Types')
plt.show()

sns.barplot(x='type', y='rating', data=anime_data)
plt.xlabel('Anime Type')
plt.ylabel('Average Rating')
plt.title('Average Rating per Anime Type')
plt.show()

sns.scatterplot(x='rating', y='members', data=anime_data)
plt.xlabel('Rating')
plt.ylabel('Number of Members')
plt.title('Rating vs. Number of Members')
plt.show()
```
